## Visualizing Waveform

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Load a sample audio file
audio_path = "./free-spoken-digit-dataset-master/recordings/0_jackson_0.wav"  # Update path
signal, sr = librosa.load(audio_path, sr=None)  # sr=None keeps original 8 kHz

# Print basic info
print(f"Sampling Rate: {sr} Hz")
print(f"Duration: {len(signal)/sr} seconds")
print(f"Signal Length: {len(signal)} samples")

# Visualize waveform
plt.figure(figsize=(10, 4))
plt.plot(np.linspace(0, len(signal)/sr, len(signal)), signal)
plt.title("Waveform of Spoken Digit")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

## Audio Signals of Digit 0

```python
import librosa
import numpy as np
import pandas as pd
from glob import glob
import os
from scipy.signal import hilbert

# Create results directory if it doesn't exist
os.makedirs("results", exist_ok=True)

recordings_path = "./free-spoken-digit-dataset-master/recordings"
audio_files = glob(os.path.join(recordings_path, "0_*.wav"))  # Only digit "0" files

features_list = []
for file in audio_files:
    signal, sr = librosa.load(file, sr=None)
    energy = np.sum(signal**2)
    if energy < 1e-6:  # Skip near-silent signals
        continue
    
    # Normalize the signal
    signal = signal / np.max(np.abs(signal))
    
    # Extract features
    zcr = np.mean(librosa.feature.zero_crossing_rate(signal)[0])
    rms = np.mean(librosa.feature.rms(y=signal)[0])
    energy = np.sum(signal**2)
    
    # Autocorrelation at lag 1
    autocorr = np.correlate(signal, signal, mode='full')
    autocorr = autocorr[len(autocorr)//2:]  # Take positive lags
    autocorr_lag1 = autocorr[1] / autocorr[0] if autocorr[0] != 0 else 0
    
    # Signal envelope using Hilbert transform
    envelope = np.abs(hilbert(signal))
    mean_envelope = np.mean(envelope)
    max_envelope = np.max(envelope)
    
    # Additional stats: duration and mean amplitude
    duration = len(signal) / sr
    mean_amplitude = np.mean(np.abs(signal))
    
    features_list.append({
        "file": os.path.basename(file),
        "digit": 0,
        "mean_zcr": zcr,
        "mean_rms": rms,
        "energy": energy,
        "autocorr_lag1": autocorr_lag1,
        "mean_envelope": mean_envelope,
        "max_envelope": max_envelope,
        "duration": duration,
        "mean_amplitude": mean_amplitude
    })

# Save to CSV
df = pd.DataFrame(features_list)
df.to_csv("./features_digit_0.csv", index=False)
print(f"Processed {len(features_list)} files for digit 0.")
```
